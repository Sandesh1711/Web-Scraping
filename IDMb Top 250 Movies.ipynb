```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
r = requests.get(url)

soup = BeautifulSoup(r.content, 'html.parser')
```

```python
#soup
```

```python
lists = soup.find('tbody', {'class':'lister-list'}).find_all('tr')
```

```python
print(lists)
len(lists)
```

```python
name = []
for a in lists:
    movie_name = a.find('td', {'class':'titleColumn'}).a.text
    name.append(movie_name)
```

```python
rating= []
for b in lists:
    rate = b.find('td',{'class': 'ratingColumn imdbRating'}).strong.text
    rating.append(rate)
    
```

```python
print(rating)
```

```python
result = {
    'Movie Name':name,
    'Rating':rating
}
```

```python
df = pd.DataFrame(result)
df.to_csv)
```
